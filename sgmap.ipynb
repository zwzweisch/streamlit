```python
import folium
```

```python
m = folium.Map(location = [1.3601710211245763, 103.81762899301569],zoom_start=11.4)
```

```python
def cm(x):
    folium.Circle(location = [x[0],x[1]],
                 radius = 300,
                 color='red',
                 fill = True,
                 popup = x[2]).add_to(m)
```

```python
df[['Latitude','Longitude','Location']].apply(lambda x: cm(x), axis=1)
```
